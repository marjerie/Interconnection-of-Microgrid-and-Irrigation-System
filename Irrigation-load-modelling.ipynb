```python
import random
import pandas as pd
with open(r'C:\Users\Marjerie\Desktop\monsoon.txt','r') as f:
    rain = f.readlines()
    rain = map(lambda s: s.strip(), rain)
rain = [float(i) for i in rain]
acre_1=200                # 300,220-280,300-350,200    :    9.5-10.5(5%),3.5-4.75(15%),7.5-8.75(10%),6.25-7(5%)
acre_2=200
pump_efficiency=0.75
motor_efficiency=0.75
land_1_load=[]
land_2_load=[]
wheat_mm=[]
rice_1_mm=[]
rice_2_mm=[]
rice_3_mm=[]
sugarcane_mm=[]
tomato_mm=[]
groundnut_mm=[]
land_1_mm=[]
land_2_mm=[]
rainpd=0
rain_per_day=[0]*153
r=0
load=[]
for day in range(365):                         # Sugarcane
    if day>=0 and day<45: 
        sugarcane_mm.append(random.uniform(6.3,7.3))
    elif day>=45 and day<120:
        sugarcane_mm.append(random.uniform(7.25,8.25))
    elif day>=120 and day<270:
        sugarcane_mm.append(random.uniform(6.2,7.2))
    elif day>=270 and day<=360:
        sugarcane_mm.append(random.uniform(6.7,7.7))
    else:
        sugarcane_mm.append(0)
for day in range(365):                    # Wheat 
    if day>=325 and day<355:
        wheat_mm.append(random.uniform(9.5,10.5))
    elif day>=355 or day<50:
        wheat_mm.append(random.uniform(5.75,6.75))
    elif day>=50 and day<90:
        wheat_mm.append(random.uniform(7.5,8.75))
    elif day>=90 and day<=120:
        wheat_mm.append(random.uniform(5.75,6.75))
    else:
        wheat_mm.append(0)
for day in range(365):                                     # Rice_1
    if day>=135 and day<165:
        rice_1_mm.append(random.uniform(4.25,5.75))
    elif day>=165 and day<225:
        rice_1_mm.append(random.uniform(6.5,7.75))
    elif day>=225 and day<265:
        rice_1_mm.append(random.uniform(10,12.5))
    elif day>=265 and day<=295:
        rice_1_mm.append(random.uniform(5,6.5))
    else:
        rice_1_mm.append(0)
for day in range(365):                               # Rice_2
    if day>=195 and day<225:
        rice_2_mm.append(random.uniform(4.25,5.75))
    elif day>=225 and day<285:
        rice_2_mm.append(random.uniform(6.5,7.75))
    elif day>=285 and day<325:
        rice_2_mm.append(random.uniform(10,12.5))
    elif day>=325 and day<=355:
        rice_2_mm.append(random.uniform(5,6.5))
    else:
        rice_2_mm.append(0)
for day in range(365):												# Rice_3
	if day>=15 and day<45:										
		rice_3_mm.append(random.uniform(4.25,5.75))
	elif day>=45 and day<105:
		rice_3_mm.append(random.uniform(6.5,7.75))
	elif day>=105 and day<145:
		rice_3_mm.append(random.uniform(10,12.5))
	elif day>=145 and day<=175:
		rice_3_mm.append(random.uniform(5,6.5))
	else:
		rice_3_mm.append(0)
for day in range(365):												# Tomato
	if day>=30 and day<60:										
		tomato_mm.append(random.uniform(1.7,2.7))
	elif day>=60 and day<90:
		tomato_mm.append(random.uniform(3.1,4.1))
	elif day>=90 and day<120:
		tomato_mm.append(random.uniform(5.0,6.0))
	elif day>=120 and day<=150:
		tomato_mm.append(random.uniform(6.0,7.0))
	elif day>=150 and day<=180:
		tomato_mm.append(random.uniform(5.5,6.5))
	else:
		tomato_mm.append(0)
for day in range(365):												# Groundnut
	if day>=195 and day<210:										
		groundnut_mm.append(random.uniform(2.0,3.0))
	elif day>=210 and day<240:
		groundnut_mm.append(random.uniform(3.4,4.4))
	elif day>=240 and day<270:
		groundnut_mm.append(random.uniform(4.8,5.8))
	elif day>=270 and day<=300:
		groundnut_mm.append(random.uniform(5.1,6.1))
	elif day>=300 and day<=330:
		groundnut_mm.append(random.uniform(1.3,2.3))
	else:
		groundnut_mm.append(0)

for n in range(len(rain)):											# Rainfall
	if rain[n]==1:
		rain[n]=random.uniform(0.05,0.25)
	elif rain[n]==2:
		rain[n]=random.uniform(0.25,1)
	elif rain[n]==3:
		rain[n]=random.uniform(1,4)
	elif rain[n]==4:
		rain[n]=random.uniform(4,16)
for n in range(len(rain)):                    
    if n%24==0 and n>0:
        rain_per_day.append(0.75*rainpd)
        rainpd=0
    if n==len(rain)-1:
        rain_per_day.append(0.75*rainpd)
    rainpd=rainpd+rain[n]
for n in range(90):
	rain_per_day.append(0)

for day in range(365):												# MM without rainfall
	land_1_mm.append(rice_1_mm[day]+wheat_mm[day])
	land_2_mm.append(sugarcane_mm[day])

for day in range(365):												# Addition of rainfall
	if day>152 and day<275:
		r=r+rain_per_day[day]
		if r>=land_1_mm[day]:
			r=r-land_1_mm[day]
			land_1_mm[day]=0
		else:
			land_1_mm[day]=land_1_mm[day]-r
			r=0
	#print land_1_mm[day]
r=0
for day in range(365):
	if day>152 and day<275:
		r=r+rain_per_day[day]
		if r>=land_2_mm[day]:
			r=r-land_2_mm[day]
			land_2_mm[day]=0
		else:
			land_2_mm[day]=land_2_mm[day]-r
			r=0
	#print land_2_mm[day]

for day in range(365):
    land_1_gpm=(0.74*land_1_mm[day]*acre_1)/(random.uniform(0.65,0.75)*6) 
    for hour in range ((day*24),(day*24)+7):
        land_1_load.append(0)
    for hour in range((day*24)+7,(day*24)+13):
        pump_efficiency=random.uniform(0.6,0.75)					# Hourly Variation
        motor_efficiency=random.uniform(0.65,0.80)
        land_1_load.append((land_1_gpm*50*0.746)/(3960*motor_efficiency*pump_efficiency))
		#print land_1_load[hour]
    for hour in range ((day*24)+13,(day*24)+24):
        land_1_load.append(0)
for day in range(365):												# Land 2
    land_2_gpm=(0.74*land_2_mm[day]*acre_2)/(random.uniform(0.65,0.75)*6)	# Daily Variation
    for hour in range ((day*24),(day*24)+7):
        land_2_load.append(0)
    for hour in range((day*24)+7,(day*24)+13):
        pump_efficiency=random.uniform(0.6,0.75)					# Hourly Variation
        motor_efficiency=random.uniform(0.65,0.80)
        land_2_load.append((land_2_gpm*50*0.746)/(3960*motor_efficiency*pump_efficiency))
		#print land_2_load[hour]
    for hour in range ((day*24)+13,(day*24)+24):
        land_2_load.append(0)
for hour in range(8760):
	load.append(land_1_load[hour]+land_2_load[hour])
	#print load[hour]

df = pd.DataFrame(data = load, columns = ['load'])
df.to_csv('irr_load.csv',index=False,encoding='utf-8')


'''for day in range(365):
	print rain_per_day[day]
print '==================='
for day in range(365):
	print land_2_mm[day]
print '==================='''
```
